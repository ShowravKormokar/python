```python
import PIL.Image as image # open image
import matplotlib.pyplot as plt # for plot to show image
import numpy as np # make array and perform array operation
```

```python
gray_img = image.open("images/hand_xray.jpg")

plt.imshow(gray_img, cmap='gray')
plt.title("Original Gray scale Image")
plt.show()
```

```python
img = np.array(gray_img)

#image size
H, W = img.shape

print(f"Resolution = {W} x {H} pixels")
```

```python
k = 128

img_th = np.zeros(img.shape)

#way: 01
# for i in range(H):
#     for j in range(W):
#         if img[i,j] > k:
#             img_th[i,j] = 255
#         else:
#             img_th[i, j] = 0
            
#Way: 02
binary_img = np.where(img > k, 255, 0)
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Gray scale Image")

plt.subplot(1, 2, 2)
# If using Way:01
# plt.imshow(img_th, cmap='gray')

# If using Way:02
plt.imshow(binary_img , cmap='gray')

plt.title("Threshold Gray Image")
plt.show()
```

```python
# Lab Task 04

for i in range(H):
    for j in range(W):
            img_th[i,j] = 255 - img[i, j]
            
            
plt.figure(figsize=(15, 8))

plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Gray Scale Image")

plt.subplot(1, 2, 2)
plt.imshow(img_th, cmap='gray')
plt.title("Negative Image")

plt.show()
```
