```python
# ================================
# Lab-1: Color Space Conversion
# ================================
```

```python
# Import required libraries
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# 1. Read an image
rgb_img = image.open("../images/test.jpeg")

# Display the original image
plt.imshow(rgb_img)
plt.title("Original RGB Image")
plt.show()
```

```python
# 2. Display the image resolution in the format: Resolution = Width  Height pixels.
width, height = rgb_img.size
print(f"Resolution = {width} x {height} pixels")
```

```python
# 3. Print image color mode
_mode = rgb_img.mode
print("Color mode is:", _mode)
```

```python
# 4. Convert RGB image to grayscale
gray_img = rgb_img.convert('L')

# Display grayscale image
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
# 5 & 6. Display original and grayscale images side by side and Label each displayed image with an appropriate title.
plt.figure(figsize=(15, 8))  # inch

plt.subplot(1, 2, 1)
plt.imshow(rgb_img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")

plt.show()
```

```python
# 7. Save the grayscale image
os.makedirs("./images/output", exist_ok=True)
gray_img.save("./images/output/grayscale.jpg")
print("Grayscale image saved successfully!")
```

```python
# ================================
# Lab-2: Brightness Adjustment of Grayscale Images
# ================================
```

```python
# 1. Read a grayscale image
gray_img = image.open("../images/test.jpeg").convert('L')

# Show original grayscale image
plt.imshow(gray_img, cmap='gray')
plt.title("Original Grayscale Image")
plt.show()
```

```python
# Convert image to numpy array
img = np.array(gray_img).astype('float32')

# Scaling factors
S1 = 1.5  # Increase brightness
S2 = 0.3  # Decrease brightness
```

```python
# 2. Increase brightness
bright_img = img * S1
bright_img = np.clip(bright_img, 0, 255).astype('uint8')
```

```python
# 3. Decrease brightness
dark_img = img * S2
dark_img = np.clip(dark_img, 0, 255).astype('uint8')
```

```python
# 4. Compare original, brighter, and darker images
plt.figure(figsize=(15, 8))

plt.subplot(1, 3, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(bright_img, cmap='gray')
plt.title("Brighter Image (S1 = 1.5)")

plt.subplot(1, 3, 3)
plt.imshow(dark_img, cmap='gray')
plt.title("Darker Image (S2 = 0.3)")

plt.show()
```
