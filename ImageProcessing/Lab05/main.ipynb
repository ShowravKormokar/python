## Lab-5: Log Transformation 
### We are tasked with writing a Python program to: 
a. Read a grayscale image. </br>
b. Apply a log transformation using the formula: s = c ⋅ log10 ( 1 + r ) s=c⋅log10​(1+r) </br>
c. Display both the original and transformed images with appropriate titles.</br>

````python
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np
````

````python
# 1. Read a grayscale image
gray_img = image.open("../images/test.jpeg").convert('L')

# Display original grayscale image
plt.imshow(gray_img, cmap='gray')
plt.title("Original Grayscale Image")

plt.show()
````

````python
# 2. Convert image to numpy array
img = np.array(gray_img).astype('float32')

# Apply log transformation
# Formula: s = c * log10(1 + r)

c = 255 / np.log10(1 + np.max(img))   # scaling constant
log_img = c * np.log10(1 + img)

# Convert back to uint8
log_img = np.clip(log_img, 0, 255).astype('uint8')
````

````python
# 3. Display original and transformed images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")


plt.subplot(1, 2, 2)
plt.imshow(log_img, cmap='gray')
plt.title("Log Transformed Image")


plt.show()
````

# ✅ Explanation (For Report / Viva)

## 🎯 Objective
To enhance the details of dark regions of a grayscale image using **log transformation**.

---

## 📌 Formula Used

\[
s = c \cdot \log_{10}(1 + r)
\]


Where:
- **r** = input pixel value  
- **s** = output pixel value  
- **c** = scaling constant  

---

## 🧠 Why Log Transformation?
- Expands dark pixel values  
- Compresses high intensity values  
- Improves visibility in darker regions  

---

## Code Explaination:
- Converts the grayscale image into a NumPy array.
- Data type is set to float32 to allow mathematical operations like logarithms without overflow or precision issues.

- Apply Log Transformation
``` c = 255 / np.log10(1 + np.max(img))```   # scaling constant
``` log_img = c * np.log10(1 + img) ```
- ```np.max(img)``` finds the maximum pixel value in the image.
- The scaling constant c ensures that transformed pixel values fit within the range 0–255.
- ```np.log10(1 + img)``` applies the log transformation to each pixel.
- Multiplying by c scales the result to the valid intensity range.

- Convert Back to Unsigned Integer
``` np.clip() 
```
- ensures all values remain between 0 and 255.
- Converts the floating-point result back to uint8 (standard image pixel format).

---

## 📊 Result
The transformed image shows **better detail in darker areas** compared to the original image.
